```python
import pandas as pd
pd.options.mode.chained_assignment = None
import matplotlib.pyplot as plt
%matplotlib inline

data = pd.read_csv('student-mat.csv')

## general knowledge
data.columns
data.shape
data.info()
```

```python
# distribute data based on Pstatus and finding out grades

df = data[['Pstatus', 'G1', 'G2', 'G3']]

df = df.groupby(['Pstatus']).sum().plot(kind='bar', stacked=True)
```

```python
# trying to find out the reason why this so much diff in grades here

# Pstatus vs famsup

df = data[['Pstatus', 'famsup']]
df.groupby('Pstatus').count().plot(kind='bar', title='Pstatus vs famsup')
```

```python
## schoolsup vs Pstatus

df = data[['Pstatus', 'schoolsup']]
df.groupby('Pstatus').count().plot(kind='bar', title='Pstatus vs schoolsup')
```

```python
## paid vs Pstatus

df = data[['paid', 'Pstatus']]
df['count'] = 1
df.groupby(['Pstatus']).sum().plot(kind='bar')
```

```python
## studytime vs Dalc / Walc -> Performance measure

# fig, ax = plt.subplots(1, 2)

df = data[['studytime', 'Pstatus']]
df.groupby(['Pstatus']).sum().plot(kind='bar')
```

```python
## travel time vs grades

df = data[['G1', 'G2', 'G3', 'traveltime']]
df['total'] = df['G1'] + df['G2'] + df['G3']
df = df[['traveltime', 'total']]
df.groupby(['traveltime']).sum().plot(title='traveltime vs grades')
```

```python
## travel time vs pstatus

df = data[['Pstatus', 'traveltime']]
df.groupby(['Pstatus']).sum().plot(kind='bar')
```
